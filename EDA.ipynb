```python
# check for missing data 
data1 = data.copy(deep=True)
print(data1.isnull().sum())

# exclude nulls
data1 = data1.dropna()
print(data1.isnull().sum())

print('Before: ', data1.shape)
data1 = data1.drop_duplicates()

print('After: ', data1.shape)
```

```python
# handle outliers
import numpy as np 
import matplotlib.pyplot as plt
from scipy import stats
z_scores = stats.zscore(data1['release_year'])
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3)
data1 = data1[filtered_entries]

print('After: ', data1.shape)
```

```python
get_ipython().getoutput('pip install scikit-learn')
from sklearn.preprocessing import StandardScaler, MinMaxScaler
# standardization
scaler = StandardScaler()

data1['release_year'] = scaler.fit_transform(data1['release_year'].values.reshape(-1,1))

# normalization
min_max_scalar = MinMaxScaler()
data1['release_year'] = min_max_scalar.fit_transform(data1['release_year'].values.reshape(-1,1))
```
